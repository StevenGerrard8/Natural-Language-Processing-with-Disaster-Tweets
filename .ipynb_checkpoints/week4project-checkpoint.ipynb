```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
import os
import keras
from tqdm.notebook import tqdm
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.regularizers import l2
from sklearn.decomposition import PCA
from keras.layers import Dense, Embedding, Bidirectional, Dropout, Flatten
from tensorflow.keras.preprocessing.text import Tokenizer
```

we can see here the except output predicted only by keyword varies a lot,there are keywords like 'aftershock' which will prvoide a 100% non-disaster prediction, however there are words like 'bombing' may implies a disaster,now let us preprocessing our data

```python
import re
import string
from nltk.corpus import stopwords
stw=set(stopwords.words('english'))
def remove_stop_words(text):
    return " ".join([word for word in text.split() if word not in stw])
def remove_spl_chars(text):
    text = re.sub('[^a-zA-Z0-9]',' ',text)
    text=re.sub('\s+',' ',text)
    return text
```

```python
train.text=train.text.apply(remove_spl_chars)
train.text=train.text.apply(remove_stop_words)
test.text=test.text.apply(remove_spl_chars)
test.text=test.text.apply(remove_stop_words)
```

```python
train_text,val_text,train_label,val_label=train_test_split(train.text,train.target,test_size=0.1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
tfidf_vectorizer = TfidfVectorizer(stop_words='english')# we use tf-idf to transform natural language to vectors after exclude all stop-words
train_text_tfidf=tfidf_vectorizer.fit_transform(train_text)
val_text_tfidf=tfidf_vectorizer.fit_transform(val_text)
```

```python
image=plt.imread('tdidf.jpg')# a hand-written tf-idf explain by me
plt.imshow(image)
```

```python
train_text_tfidf#we can see that we have 19794 rows which means that this is a total number of 6851x19706  non-stop words in the trainset 
```

```python
train_text_tfidf=train_text_tfidf.toarray()#transform sparse matrix to dense matrix
val_text_tfidf=val_text_tfidf.toarray()
```

now let us build and train our model

```python
max_features=train_text_tfidf.shape[1]
maxlen=10000
train_text_pad=pad_sequences(train_text_tfidf, maxlen=maxlen)
val_text_pad=pad_sequences(val_text_tfidf, maxlen=maxlen)
```

```python
model = Sequential()
model.add(Embedding(input_dim=max_features, output_dim=128))
model.add(LSTM(units=64))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(train_text_pad, train_label, epochs=10, batch_size=32, validation_data=(val_text_pad, val_label))
```

As you can see from the result above,I manually shut down the model because training for half of the features on the tfidf(10000/19598) takes nearly half an hour(1680s/60s=28 min) for one epoch! And the accuracy is so bad.Why this happens? the reason behind this is that the tf-idf matrix I build is so sparse and the majority entry of the matrix is 0,which will make the model generate way too many coeffcients to predict based on that,so it greatly enlarge the training time and make the model overfit,so the accuracy is bad. 

to address this problem, Another vectorize method have to be done.

we can see that the training time is significantly lower, now let us see how to fine tune our model to reach a greater accuracy

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(
        input_dim=200,
        output_dim=128,
        # Use masking to handle the variable sequence lengths
        mask_zero=True),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(100)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
pca=PCA(n_components=50,svd_solver='randomized')
train_pca=pca.fit_transform(train_text_tfidf)
val_pca=pca.fit_transform(val_text_tfidf)
```

```python
model.fit(train_pca, train_label, epochs=10, batch_size=16, validation_data=(val_pca, val_label))
```

<b>PCA significantly increase the speed, however the accuracy is still not good,It looks like we need a new way for vectorization<b>

I will try Glove a.k.a Global vector for vectorization.Sp what is Glove?I will show a hand-written analising on it.
by the way, if you are interested, there is the original paper:   https://aclanthology.org/D14-1162/

```python
plt.imshow(np.rot90(plt.imread('glove.jpg')))
```

```python
X_train = train.text.values.reshape(len(train),)
X_test = test['text'].values.reshape(len(test),)
total_tweets = np.concatenate((X_train, X_test))
tokenizer = Tokenizer()#first step ,tokenize texts
tokenizer.fit_on_texts(total_tweets)
```

```python
X_train_token = tokenizer.texts_to_sequences(X_train)
X_test_token = tokenizer.texts_to_sequences(X_test)
```

```python
word_index=tokenizer.word_index
len(word_index)# the total number of words after cleaning
```

```python
X_train_pad = pad_sequences(X_train_token, maxlen=max(len(x.split()) for x in total_tweets), padding='post')
X_test_pad = pad_sequences(X_test_token, maxlen=max(len(x.split()) for x in total_tweets), padding='post')
```

```python
embedding_index = {}
with open('glove.6B.300d.txt', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs#we load the pretraines model
```

```python
train_embd=np.zeros((len(word_index)+1,300))
for word,i in word_index.items():
    vec=embedding_index.get(word)
    if vec is not None:
        train_embd[i]=vec#for every word in our data, if it exists in the pre-trained glove model,we get their pre_train-ed vectors
```

```python
train_embd
```

```python
model = Sequential()
layer=Embedding(input_dim=len(word_index)+1, output_dim=200,trainable=False)
model.add(layer)
model.add(LSTM(units=100, return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(X_train_pad, train.target, epochs=50, batch_size=32, validation_split=0.1)
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Plot training and validation loss
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

we can see here that the val_loss just keeps rising ,indicating that there is potential overfitting,so let us change our model

```python
model_LS = Sequential()
model_LS.add(Embedding((len(word_index)+1), 200))
model_LS.add(Bidirectional(LSTM(100)))
model_LS.add(Dropout(0.3))
model_LS.add(Dense(100, activation='relu', kernel_regularizer=l2(0.01)))#we add regulation to our model
model_LS.add(Dropout(0.3))
model_LS.add(Dense(1, activation='sigmoid'))
model_LS.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
from keras.callbacks import EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss',  # Metric to monitor
                               patience=2 ,        # Number of epochs with no improvement after which training will be stopped
                               restore_best_weights=True) # early_stopping our model to prevent overfitting
```

```python
history=model_LS.fit(X_train_pad, train.target, epochs=60, batch_size=21, validation_split=0.2,callbacks=[early_stopping])
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Plot training and validation loss
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

<b>it seems that our model converges so fast and it still turns to overfit<b>

```python
from keras.optimizers import Adam
model_LS = Sequential()
model_LS.add(Embedding((len(word_index)+1), 200))
model_LS.add(Bidirectional(LSTM(100)))
model_LS.add(Dropout(0.3))
model_LS.add(Dense(100, activation='relu', kernel_regularizer=l2(0.01)))
model_LS.add(Dropout(0.3))
model_LS.add(Dense(1, activation='sigmoid'))
opt = Adam(learning_rate=0.00003)#turn down the learning rate
model_LS.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
early_stopping = EarlyStopping(monitor='val_loss',  # Metric to monitor
                               patience=3 ,        # Number of epochs with no improvement after which training will be stopped
                               restore_best_weights=True) # early_stopping our model to prevent overfitting
```

```python
history=model_LS.fit(X_train_pad, train.target, epochs=60, batch_size=21, validation_split=0.2,callbacks=[early_stopping])
```

```python
train_loss = history.history['loss']
val_loss = history.history['val_loss']

# Plot training and validation loss
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Training Loss')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```

this time, we got our best result

```python
pred=model.predict(X_test_pad)
sub=pd.read_csv('submit.csv')
sub['target']=list(map(lambda x: 0 if x < 0.5 else 1, pred))
sub.to_csv('submit.csv', index=False)
```

Result, Analysis
</b></font>

    I tried possible learning rate from 0.000001-0.0001, different kinds of regulation and different drop out value and several encodering method to vectorize texts. The best result of those are shown above, I reach a val/test accuracy of 74%. For individual hyperparameters, the effect of learning rate has shown above, when the lr is too high,the model overshoot so it actually reach the best val loss and accuracy in the first epoch,When it is too low ,the model converges so low, it need too many epoches for it to converge.For the drop out value,too-high dropout rates can lead to underfitting, where the model is unable to capture the underlying patterns in the data effectively. This happens because dropout removes too many connections during training, preventing the network from learning meaningful representations from the data. When it is too low, the affect it put on the model is too slim to even notice.
